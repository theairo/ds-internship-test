```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt

log_path = "logs/events.out.tfevents.1763813100.Kolya.4948.0"
ea = event_accumulator.EventAccumulator(log_path)
ea.Reload()

# Pick the tags you want
train_tag = 'train/loss'
eval_tag = 'eval/loss'

# Load events
train_events = ea.Scalars(train_tag)
eval_events = ea.Scalars(eval_tag)

# Extract steps and values
train_steps = [e.step for e in train_events]
train_values = [e.value for e in train_events]

eval_steps = [e.step for e in eval_events]
eval_values = [e.value for e in eval_events]

# Plot
plt.figure(figsize=(8,5))
plt.plot(train_steps, train_values, label='Train Loss')
plt.plot(eval_steps, eval_values, label='Eval Loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training and Evaluation Loss')
plt.legend()
plt.grid(True)
plt.show()
```
